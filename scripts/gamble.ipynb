```python
import random 

start_income = 100
money = start_income
bet = 1 
games = 1
bet_one = [start_income]
def bet_single(money, bet, games):
    money_list = []
    games = 0 
    while money > 0:
        if random.random() < 0.5:
            money -= bet
        else:
            money += bet
        games += 1 
        money_list.append(money)
    return money_list, games
def bet_all(money, bet):
    money_list = []
    games = 0 
    while money > 0:
        if random.random() < 0.5:
            money = 0 
        else:
            money = money * 2 
        games += 1 
        money_list.append(money)
    return money_list,games
def martingale(money):
    money_list = []
    games = 0 
    bet = 1 
    while money > 0:
        if random.random() < 0.5:
            money -= bet 
            bet = bet * 2
        else:
            money += bet
            bet = 1 
        games += 1 
        money_list.append(money)
    return money_list,games
        
bet_one,number_of_games_bet_one = bet_single(money, bet, games)
bet_all,number_of_games_bet_all = bet_all(money,bet)
martingale_list, number_of_games_martingale = martingale(money)
games_bet_one = [i for i in range(number_of_games_bet_one)]
games_bet_all = [i for i in range(number_of_games_bet_all)]
games_martingale = [i for i in range(number_of_games_martingale)]

```

```python
import matplotlib.pyplot as plt 
fig, ax = plt.subplots()
line1, = ax.plot(games_bet_one,bet_one,label='Bet One')
line2, = ax.plot(games_bet_all, bet_all, label="Bet All")
line3, = ax.plot(games_martingale,martingale_list,label="Martingale")
ax.legend(handles=[line1,line2,line3])
plt.show()
    
```
